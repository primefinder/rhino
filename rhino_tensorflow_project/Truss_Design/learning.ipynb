```python
import tensorflow as tf
import numpy as np
import csv

xy = np.loadtxt('output.csv', delimiter=',', dtype=np.float32)
#print(xy)

loop = xy[:,0:1]
x_data = xy[:,1:4]
y_data = xy[:,4:5]

#print(loop)
#print(x_data)
print(x_data.shape)
#print(y_data)
print(y_data.shape)

######################################################################
X = tf.placeholder(tf.float32, shape=[None, 3])
Y = tf.placeholder(tf.float32, shape=[None, 1])

W1 = tf.Variable(tf.random_normal([3,16]), name='weight1')
b1 = tf.Variable(tf.random_normal([16]), name='bias1')
L1 = tf.nn.relu(tf.matmul(X, W1) + b1)

W2 = tf.Variable(tf.random_normal([16,8]), name='weight2')
b2 = tf.Variable(tf.random_normal([8]), name='bias2')
L2 = tf.nn.relu(tf.matmul(L1, W2) + b2)

W3 = tf.Variable(tf.random_normal([8,4]), name='weight3')
b3 = tf.Variable(tf.random_normal([4]), name='bias3')
L3 = tf.nn.relu(tf.matmul(L2, W3) + b3)

W4 = tf.Variable(tf.random_normal([4,2]), name='weight4')
b4 = tf.Variable(tf.random_normal([2]), name='bias4')
L4 = tf.nn.relu(tf.matmul(L3, W4) + b4)

W5 = tf.Variable(tf.random_normal([2,1]), name='weight5')
b5 = tf.Variable(tf.random_normal([1]), name='bias5')

hy = tf.matmul(L4, W5) + b5

cost = tf.reduce_mean(tf.square(hy - Y))

optimizer = tf.train.AdamOptimizer(learning_rate=0.001)
train = optimizer.minimize(cost)
######################################################################

sess = tf.Session()
sess.run(tf.global_variables_initializer())

for step in range(20001):
    W1_1, W2_1, W3_1, W4_1, W5_1, b1_1, b2_1, b3_1, b4_1, b5_1, cost_val, hy_val, _ = sess.run([W1, W2, W3, W4, W5, b1, b2, b3, b4, b5, cost, hy, train], feed_dict={X: x_data, Y: y_data})
    if step % 100 == 0:
        print(step, "Cost: ", cost_val)       

sess.close()
```
